```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Lineplot
```

```python
x=np.arange(1,10+1,1)
x
```

```python
y=2*x
y
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,color="red")
plt.title("Lineplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,color="red",linestyle="-.",linewidth=2)
plt.title("Lineplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
z=3*x
z
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,color="red",linestyle="--",linewidth=2)
plt.plot(x,z,color="green",linestyle=":",linewidth=3)
plt.title("Lineplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color="red",linestyle="--",linewidth=2)
plt.grid()
plt.subplot(1,2,2)
plt.plot(x,z,color="green",linestyle=":",linewidth=3)
plt.grid()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color="red",linestyle="--",linewidth=2)
plt.grid()
plt.subplot(2,1,2)
plt.plot(x,z,color="green",linestyle=":",linewidth=3)
plt.grid()
plt.show()
```

```python
#Barplot
```

```python
student={"Alex":100,"Zion":90,"John":80,"Michael":70,"Samuel":60}
print(student)
print(student.keys())
print(student.values())
print(student.items())
```

```python
names=list(student.keys())
print(names)
```

```python
marks=list(student.values())
print(marks)
```

```python
plt.figure(figsize=(6,4))
plt.bar(names,marks,color="magenta")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(names,marks,color="green")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(names,marks,color="red")
plt.ylabel("Names")
plt.xlabel("Marks")
plt.grid()
plt.show()
```

```python
import random
```

```python
x=random.randint(1,100)
x
```

```python
x=random.sample(range(1,10+1),10)
random.seed(0)
x
```

```python
y=random.sample(range(11,20+1,1),10)
random.seed(0)
y
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color="brown")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color="orange",marker="*",s=100)
plt.title("Scatterplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
z=random.sample(range(21,30+1,1),10)
random.seed(0)
z
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color="red",marker="*",s=100)
plt.scatter(x,z,color="green",marker="o",s=100)
plt.title("Scatterplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,color="red",marker="*",s=100)
plt.grid()
plt.subplot(1,2,2)
plt.scatter(x,z,color="green",marker="o",s=100)
plt.grid()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,y,color="red",marker="*",s=100)
plt.grid()
plt.subplot(2,1,2)
plt.scatter(x,z,color="green",marker="o",s=100)
plt.grid()
```

```python
#Piechart
```

```python
mncs=["Microsoft","Amazon","Meta","Google","Apple"]
quantity=[30,20,20,15,5]
```

```python
plt.figure(figsize=(6,4))
plt.pie(quantity,labels=mncs)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.pie(quantity,labels=mncs,autopct="%0.1f%%",colors=["red","orange","yellow","blue","green"])
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.pie(quantity,labels=mncs,autopct="%0.1f%%",colors=["red","orange","yellow","blue","green"],radius=1)
plt.pie([1],colors=["white"],radius=0.5)
plt.show()
```
